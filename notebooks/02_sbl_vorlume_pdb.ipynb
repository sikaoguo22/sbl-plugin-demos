### VMD plugin
The plugin is available in VMD under Extensions → SBL. The interface has two main areas: an input area for selecting the molecular structure and configuring calculation parameters, and an output area for inspecting results. After clicking the Run button, the plugin launches sbl-vorlume-pdb.exe with the chosen settings in the background, then automatically runs a post‑analysis script to prepare the outputs. Once the computation and post‑analysis are complete, VMD displays the molecular surface directly in the viewer together with the computed volume and surface area values in the GUI.

### VMD plugin
The plugin is available in VMD under Extensions → SBL. The interface has two main areas: an input area for selecting the molecular structure and configuring calculation parameters, and an output area for inspecting results. After clicking the Run button, the plugin launches sbl-vorlume-pdb.exe with the chosen settings in the background, then automatically runs a post‑analysis script to prepare the outputs. Once the computation and post‑analysis are complete, VMD displays the molecular surface directly in the viewer together with the computed volume and surface area values in the GUI.

![Open sbl_vorlume_pdb VMD plugin](../figs/vorlume_pdb_open_plugin.png)

#### Prepare the work directory and molecular structure (pdb file)
```bash
mkdir -p ~/demo/vorlume-pdb/data 
cd ~/demo/vorlume-pdb/data
sbl-pdb-fetch.py -f 1CRN -c
```

### PyMOL plugin
Note: The PyMOL is required to built with PyQt6 support for the SBL PyMOL plugins to work correctly. PyMOL with PyQt5 is not supported.

Run `sbl_help` in the PyMOL command line to list all available SBL plugin commands, then run `sbl_vorlume_pdb` to launch this plugin.

The usage of this plugin is the same as in VMD.

### Web plugin
Run `sbl-web-plugins-launcher` to start the SBL web plugins launcher locally, then open `http://127.0.0.1:8700` in a web browser to access it. Click the `Open` button to launch the corresponding plugin in a new tab.

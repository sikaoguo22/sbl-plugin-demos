### PyMOL plugin
Note: The PyMOL is required to built with PyQt6 support for the SBL PyMOL plugins to work correctly. PyMOL with PyQt5 is not supported.

Run `sbl_help` in the PyMOL command line to list all available SBL plugin commands, then run `sbl_alphafold_dbrun` to launch this plugin.

The usage of this plugin is the same as in VMD.
